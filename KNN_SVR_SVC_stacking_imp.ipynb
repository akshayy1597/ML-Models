```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/winequality-red.csv')
```

```python
df
```

```python
x = df.drop(columns = 'quality')
x
```

```python
y = df['quality']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=20)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

```python
param = {
    'n_neighbors': [3,5,7,9,12,13,15,17,21],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [10,15,20,25,30,35,40,45],
    'p': [1,2],
    'weights': ['uniform', 'distance']

}
```

```python
grid_cv = GridSearchCV(knn, param_grid=param)
```

```python
grid_cv.fit(x_train, y_train)
```

```python
grid_cv.best_params_ #for checking with 17 neighbors
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df2 = pd.read_csv('/content/winequality-red.csv')
```

```python
df2
```

```python
x = df2.drop(columns='quality')
y = df2.quality
```

```python
train, val_train, test, val_test = train_test_split(x,y,test_size = .50, random_state = 30) ##50records for train and 50for validation
##50 divided into 40,10: 40=x_train, 10=x_test
##this 40 is given to KNN n SVC
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, test, random_state = 30 , test_size = .20)
```

```python
 knn = KNeighborsClassifier()
 knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

```python
svc = SVC()
svc.fit(x_train, y_train)
```

```python
svc.score(x_test, y_test)
```

```python
prediction_knn = knn.predict(val_train)
prediction_svc = svc.predict(val_train)
```

```python
input3 = np.column_stack((prediction_knn, prediction_svc))
```

```python
output = val_test
output
```

```python
pd.DataFrame(input3)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(input3, output)
```

100% of data divided into 50% 50%.

one 50% part is futher duvided into 40% n 10%, this 10% is for model validation

40% given to knn n svm for training

a new model is created with the 50% left over data nd the data from knn n svm to make predictions

nd that prediction is further sent into random forest model

```python
knn_output = knn.predict(x_test)
svc_output = svc.predict(x_test)
```

```python
output_stack1 = np.column_stack((knn_output, svc_output))
```

```python
rf.predict(output_stack1)
```

```python
rf.score(output_stack1, y_test)
```
